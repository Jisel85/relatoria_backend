```python
import time
import concurrent.futures
import pandas as pd

inicio = 0 # o empezar de 0 para otra cosa
fin = 41712
tamao_chunk = 1000

df = pd.DataFrame()

chunks = []
for i in range(inicio, fin + 1, tamao_chunk):
    chunks.append((i, min(fin, i + tamao_chunk - 1)))

def process(row):
    return row['metadata']

for inicio, fin in chunks:
    print('iniciando', inicio, fin)
    resultados = index.fetch(ids=list(map(str, range(inicio, fin))))
    print('datos obtenidos', inicio, fin)

    with concurrent.futures.ThreadPoolExecutor() as executor:
        lista = executor.map(process, list(resultados['vectors'].values()))
        otro_df = pd.DataFrame(lista)
        df = pd.concat([df, otro_df])
    print('datos procesados', inicio, fin)
```

```python
import sys
from tqdm.auto import tqdm

# we will use batches of 64
batch_size=64

def insertar_index(data, limit = None):
    if limit:
        for el in data:
            if len(el[2]['summary_extract']) > limit:
                print(el[0], len(el[2]['summary_extract']))
            el[2]['summary_extract'] = el[2]['summary_extract'][:limit]
    try: 
        index.upsert(data)
        return True
    except:
        return False

for i in tqdm(range(0, len(df1), batch_size)):

    # find end of batch
    i_end = min(i+batch_size, len(df1))
    # extract batch
    batch = df1[i:i_end]
    # generate embeddings for batch
    ids = batch['ids']
    emb = batch['embeddings']
    metadata = batch.drop(['ids','embeddings','text'],axis=1).to_dict('records')
    

    # add all to upsert list
    to_upsert = list(zip(ids, emb, metadata))

    # update/insert these records to pinecone
    result = insertar_index(to_upsert)
    if not result:
        result = insertar_index(to_upsert, 60000)
    if not result:
        result = insertar_index(to_upsert, 55000)
    if not result:
        result = insertar_index(to_upsert, 50000)
    if not result:
        result = insertar_index(to_upsert, 45000)
    if not result:
        result = insertar_index(to_upsert, 40000)
    if not result:
        result = insertar_index(to_upsert, 35000)
    if not result:
        result = insertar_index(to_upsert, 30000)
# check that we have all vectors in index
index.describe_index_stats()
```
